```python
import pandas as pd
import openap
import matplotlib.pyplot as plt

pd.options.display.max_columns = 100
```

```python
flight = pd.read_csv("data/flight_a320.csv").query("ALTI_STD_FT>500")
flight.head()
```

```python
fuelflow = openap.FuelFlow("a320")


acc = (
    flight.TRUE_AIR_SPD_KT.diff().bfill()
    * openap.aero.kts
    / flight.FLIGHT_TIME.diff().bfill()
)

ff_estimate = fuelflow.enroute(
    mass=flight.MASS_KG,
    tas=flight.TRUE_AIR_SPD_KT,
    alt=flight.ALTI_STD_FT,
    vs=flight.VERT_SPD_FTMN,
)

plt.plot(
    flight.FLIGHT_TIME,
    flight.FUEL_FLOW_KGH * 2 / 3600,
    lw=1,
    label="actual (kg/h)",
)
plt.plot(
    flight.FLIGHT_TIME,
    ff_estimate,
    alpha=0.8,
    lw=1,
    label="estimate (kg/h)",
)
plt.legend()
```
